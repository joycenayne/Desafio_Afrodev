#### Equipes


##### Quantos times por esporte cada país tem

```python
equipes
```

```python
paises = list(equipes["NOC"].unique())
list_times_pais = []
for pais in paises:

    dict_times_pais = {
        "País":'Australia',
        "Times": []
    }
    data_times_pais = equipes[equipes.NOC == 'Australia'].groupby(by="Discipline").count()['Event']
    esportes =  list(data_times_pais.index)
    qtde_times = list(data_times_pais.values)

    for i in range(0, data_times_pais.shape[0]):
        dict_times = {}
        dict_times["Esporte"] = esportes[i]
        dict_times["Quantidade"] = qtde_times[i]
        dict_times_pais["Times"].append(dict_times)

list_times_pais.append(dict_times_pais)
```
